> **Ejercicio**: Defina una función que convierta grados centígrados en grados Farenheit.

```python
#<SOL>
# Grados Fahrenheit = (grados centígrados × 9/5) +32
def celsiusAFarenheit(celsius):
  gradosFarenheit = celsius * 9 / 5 + 32
  return gradosFarenheit

# Ejemplo para comprobar el resultado
gradosCentigrados = 21
gradosFarenheit = celsiusAFarenheit(gradosCentigrados)
print(gradosCentigrados, 'grados centigrados son', gradosFarenheit, 'grados Farenheit')

#</SOL>
```

> **Ejercicio**: Diseñe una función que reciba una cadena y devuelva True si la cadena empieza por minúscula y False en caso contrario.

```python
#<SOL>
def comprobarMinuscula(cadena):
  if cadena[0].islower():   # if cadena[0] >= 'a' and cadena[0] <= 'z':
    respuesta = True
  else:
    respuesta = False
  return respuesta

cadenaDePrueba = "Fasd asd"
resultado = comprobarMinuscula(cadenaDePrueba)
if resultado:
  print('La cadena empieza por minúscula')
else:
  print('La cadena no empieza por minúscula')

#</SOL>
```

```python
def comprobarCadena(cadena):
  if cadena[0].islower():   # if cadena[0] >= 'a' and cadena[0] <= 'z':
    respuesta = True
  elif cadena[0].isupper():
    respuesta = False
  else:
    respuesta = None
  return respuesta

cadenaDePrueba = "3Fasd asd"
resultado = comprobarCadena(cadenaDePrueba)
if resultado == True:
  print('La cadena empieza por minúscula')
elif resultado == False:
  print('La cadena empieza por mayúscula')
else:
  print('No empieza por una letra')
```

### 1.4 Especificar parámetros con valores por defecto (Extra!)

Es posible especificar valores por defecto en los parámetros de una función. El valor por defecto se usará cuando en la llamada a la función no se haya proporcionado el correspondiente argumento.

```python
def menu(bebida, entrante, postre='Tiramisú'):
    return {'bebida': bebida, 'entrante': entrante, 'postre': postre}
```

```python
# Hacemos uso del valor por defecto del parámetro "postre"
menu('Vino', 'Ensalada')
```

```python
# "Sobreescribimos" el valor de "postre" especificando uno concreto
menu('Agua', 'Revuelto de setas', 'Helado')
```

Los valores por defecto en los parámetros se calculan cuando se **define** la función, no cuando se **ejecuta**.

Escribe una función que tome un número como parámetro y devuelva True si es divisible por 3 y False en caso contrario. Luego, solicita al usuario un número e imprime si es divisible por 3 utilizando la función

```python
def divisible_3(n):
  if n % 3 == 0:
    return True
  else:
    return False

# Programa principal
numero = int(input('Introduce un número: '))

if divisible_3(numero):
  print('El número', numero, 'es divisible por 3')
else:
  print('El número', numero, 'no es divisible por 3')
```

Escribe una función que tome un número como parámetro y calcule la suma de todos los números pares desde 1 hasta ese número. Utiliza un bucle for dentro de la función para realizar la suma. Luego, solicita al usuario un número e imprime la suma de los números pares utilizando la función.

```python

def suma_pares(num):

  suma = 0
  for sumando in range(2, num+1, 2):
    suma = suma + sumando
  return suma

numero = int(input('Introduce un número: '))

resultado = suma_pares(numero)
print(resultado)
print(suma_pares(numero))

print(suma_pares(2))
print(suma_pares(10))
print(suma_pares(200))
```
